# Toronto Bikeshare Project  - Data Cleaning

The Toronto Bikeshare Project aims to use open source data from [Toronto's Bikeshare Program](https://bikesharetoronto.com/pricing/?gclid=Cj0KCQiAz9ieBhCIARIsACB0oGIMFeM961gRm3NjO-8os2Q4FqV7dm4YVdArkEpCOa5caaQnc2nRWe4aAhUuEALw_wcB) and conduct Time Series as well as Network Analysis on it. The aim is to predict ridership changes on different temporal intervals as well as look at how ridership occurs for a potential bike from 1 station over a few days. These kind of analyses can assist with operational decisions for the organisation such as when to potentially increase bike availability in certain stops based on day or week or where additional stops can be added based on usage patterns. 

In this notebook the data will be loaded and cleaned for use in further exploration and analysis later. Let's start by loading the data directly using a Python API from the [Toronto Toronto Open Data website](https://open.toronto.ca/dataset/bike-share-toronto-ridership-data/). The API is copied and used below.

```python
# package
import pandas as pd
```

```python
# Create an empty list to store all the files.
dataframe = []

# Import glob to read all the csv files
import glob

# Create a for loop to read all the csv files in the folder
for file in glob.glob('data/*.csv'):
    
    # Read each file as a csv 
    data = pd.read_csv(file)
    
    # Append each file to the list
    dataframe.append(data)
```

The dataframe can now be created by concatenating the data in the list.

```python
# Create dataframe by concatenating the data in the list. 
tor_biking = pd.concat(dataframe)
```

The first 10 rows of the dataframe can be seen below. 

```python
tor_biking.head()
```

The data contains the following columns. 

|Column Name   |  Column Description |   
|---|---|
|  Trip Id | Unique identification number for the trip  |
|  Trip Duration | Duration of the trip in seconds  |
|  Start Station Id |  Unique identification number for the station where the trip started|
|  Start Time	 | Date and time in hours and minutes for the start time of the trip  |
|  Start Station Name	 |  Name of station where the trip started |
|  End Station Id	 | Unique identification number for the station where the trip ended  |
|  End Time	|  Date and time in hours and minutes for the end time of the trip |
|  End Station Name |  Name of station where the trip ended |
|  Bike Id	 | Unique identification number for the bike used  |
|  User Type | Membership type of user who undertook the trip  |

#### Changing  `Trip Duration` to Minutes

The `Trip Duration` column is in seconds which makes data interpretation unnecessarily complex. Therefore, as [previously](#FloatDataType) discussed, the column will be changed to minutes as most trips are likely to be below 45 minutes. This is because a surcharge is added for trips exceeding 30 or 45 minutes depending on [membership type](https://bikesharetoronto.com/pricing/) (check bottom of website page for pricing surcharges). First, let's check the statistics of the current `Trip Duration` column.

```python
# Look at the statistics of the column
tor_biking['Trip  Duration'].describe()
```

It seems that there is an extra space in the name of the column. This will be fixed below. 

```python
tor_biking.rename(columns={'Trip  Duration':'Trip Duration'}, inplace=True)
```

Looking at the statistics above the mean is ~969 seconds with the median being ~720 seconds. The columns will be converted below and the statistics rechecked to ensure that there are no unexpected changes other than the values appearing in minutes.

```python
#Convert the column to minutes by dividing by 60
tor_biking['Trip Duration'] = tor_biking['Trip Duration']/60
```

Let's ensure the changes have been made successfully.

```python
tor_biking.head()
```

The numbers are looking reasonable now, though this will be further checked using statistics below. 

```python
tor_biking['Trip Duration'].describe()
```

The statistics above show the average trip being ~16.15 minutes long and the median trip being 12 minutes long. These equate to the values seen previously (969 and 720). Therefore the conversion is successful.
Note that `Trip duration` is now of float type with a very high decimal point allowance. This will be decreased to clean the data further. 

```python
tor_biking['Trip Duration'] = tor_biking['Trip Duration'].round(decimals=2)
```

Let's ensure that the changes took place appropriately. 

```python
tor_biking.head()
```

As we can see the changes seem appropriate and the `Trip Duration` column only has 2 decimal place values. We can confirm below that the data types are all correct.

Finally, it would be good to change the column names to make it easier to auto-complete code. Specifically all the column names will be changed to include `_` instead of space and turn all upper case into lower case. This is undertaken below.

```python
# Create an empty list which will hold the updated names after transformation.
new_name = []

# Create a loop that goes over every column name
for column in tor_biking.columns:
    
    # Use .lower to convert the uppercase letters to lowercase
    column_lower = column.lower()
    
    # Replace space within names to underscore
    column_correct = column_lower.replace(" ", "_")
    
    # append each updated column name in the list
    new_name.append(column_correct)
```

Confirm the list has the correct names. 

```python
new_name
```

The new names look appropriate. Let's rename the columns in the dataset below. 

```python
# create a for loop that takes the each column name and number and replaces them using the 
# new_name list created earlier. 
for number, column in enumerate(tor_biking.columns):
    tor_biking.rename(columns={column:new_name[number]}, inplace=True)
```
